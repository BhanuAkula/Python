```python
#Task 1#
x=np.linspace(0,20,200)

fig = plt.figure()
plt.style.use('bmh')
plt.plot(x, np.sin(x), 'r--')
plt.plot(x, np.cos(x), 'k-.')

plt.title('Sine and Cosin Curves')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend(('sinx(x)', 'cos(x)'))
plt.show()
fig.savefig('sine_cosine_figure.png')
```

```python
# Task 3 #
x = np.linspace(0,10,20)
y = np.sin(x)


rng = np.random.RandomState(421)
colors = rng.rand(20)
sizes = 1000 * rng.rand(20)

fig = plt.figure()

plt.scatter(x,y,c=colors, s=sizes, cmap='tab20c')
plt.colorbar()
plt.show()

fig.savefig('dots_figure.png')
```
