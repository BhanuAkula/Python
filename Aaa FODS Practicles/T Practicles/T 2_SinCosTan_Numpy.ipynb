Define the sigmoid function (𝑓=11+𝑒−𝑥) using a Lambda function and plot the curve in the range of 𝑥=[−10,10].

```python
#define lambda functionn for sigmoid
'''
    Calulates the sigmoid value for values passed as x
'''
f = lambda x : 1/(1+np.exp(-x))

# get an array of values between -10 to 10
x = np.linspace(-10,10)

#calculating the value for y when x is in the range of -10 to 10 
y = f(x)

#plot the resulting array
plt.plot(x, y)

#plt.title('')
plt.xlabel('[-10,10]')
plt.ylabel('sigmoid')

plt.tight_layout()
plt.show()
```

Define the same function using the key words def and return. Evaluate the function value when 𝑥=0by calling the function you have defined.

```python
def calcSigmoid(x):
    '''
    Calulates the sigmoid value for values passed as x
    '''
    f = 1/(1+np.exp(-x))
    return f
```

```python
# get an array of values between -10 to 10
new_x = np.linspace(-10,10)

#calculating the value for y when x is in the range of -10 to 10 
new_y = calcSigmoid(new_x)

print (new_y)
```

# Task 11  - Limits 

```python
'''S contains the definition of some special values, 
such as positive  infinity, negative  infinity, and so on.'''
from sympy import Limit, Symbol, S
```

```python
#needs to be defined as a symbol
x = Symbol('x')

#limit(function definition, variable, approaching value)
# find the limit of 1/x when x is approaching Infinity
l = Limit(1/x, x, S.Infinity)
print(l)
l.doit()
```

dir =  '-' refers to X approaching from the left 

```python
l = Limit(1/x, x, 0)
print(l)
l.doit()
```

# Task 12 - Derivatives

```python
from sympy import Symbol, Derivative
```

```python
x = Symbol('x')
f = 100*x**2 + 4*x + 9 
print(f)
```

```python
#calculate Derivative of f for a change in x
Derivative(f,x).doit()
```

```python
#calculating the derivative at a specific value of X
df = Derivative(f,x)
print(df)
#derivative when x  = 1
df.subs({x:1}).doit()
#Another method. => Derivative(f,x).subs({x:1}).doit()
```
