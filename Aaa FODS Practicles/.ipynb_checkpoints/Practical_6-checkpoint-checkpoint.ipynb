```python
#Task 1#
def probability(space, event):
    '''define the probability for a uniform distribution'''
    '''P(E) = n(E)/n(S), n(E) and n(S) are the cardinality of the sets E, the event, and S, the sample space'''
    prob = len(event)/len(space)
    return prob
```

```python
'''The aim of this function is to check if a number if an even number'''
def check_even(number):
    for i in range(2, number):
        if number % 2 !=0:
            return False
    return True
```

```python
if __name__=='__main__':
    space = range(2,50)
    evens = []
    for j in space:
        if check_even(j):
            evens.append(j)
    p = probability(space, evens) 
    
    print('Sample space: {0}'.format(space))
    print('Event: {0}'.format(evens))
    print('Probability of even numbers: {0:.5f}'.format(p))
```

```python
from scipy.stats import bernoulli


r = bernoulli.rvs(size=5000, p=0.2)

ax= sns.distplot(r,
                 kde=False,
                 color="black",
                 hist_kws={"linewidth": 15,'alpha':0.6})
ax.set(xlabel='Bernoulli', ylabel='Frequency')
```

```python
from scipy.stats import binom
r = binom.rvs(size = 10000, p=0.6, n=20)
ax= sns.distplot(r,
                 kde=False,
                 color="blue",
                 hist_kws={"linewidth": 15,'alpha':0.5})
ax.set(xlabel='Bernoulli', ylabel='Frequency')
```

```python
from scipy.stats import poisson
r = poisson.rvs(mu=0.2,size=5000)

ax = sns.distplot(r,
                  kde=False,
                  color='red',
                  hist_kws={"linewidth": 15,'alpha':0.5})
ax.set(xlabel='Poisson', ylabel='Frequency')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu');
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,y);
```

```python
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14 + 18] * rng.rand(1000,2)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim);
```

```python
ynew = model.predict(Xnew)

# Task 7 #

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim);
```

```python
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```

```python
# Task 8.1 #
from sklearn import datasets
wine = datasets.load_wine()
print("Feature Names: ", wine.feature_names)
print("Labels: ", wine.target_names)
```

```python
# Task 8.2 #
wine.data.shape
```

```python
wine.target
```

```python
# Task 8.3 #

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.4, random_state=1)
```

```python
# Task 8.4 & 8.5 #

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred= gnb.predict(X_test)
```

```python
# Task 8.6 #
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
# Task 9 #
class Student:
    """contents of the class"""
    def __init__(self, pModule, pName, pMark):
        self.module = pModule
        self.name = pName
        self.mark = pMark
        print('Creating Student Object')    
        
    def __str__(self):
        return "Module = %s, Name = %s, Mark = %d" %(self.module, self.name, self.mark)
    
    def totalMark(self):
        prompt = '\nEnter the mark of the first coursework %s:' %(self.name)
        mark1 = input(prompt)
        prompt = '\nEnter the mark of the second coursework %s:' %(self.name)
        mark2 = input(prompt)
        self.mark = int(mark1)+int(mark2)
        return self.mark
    
```
